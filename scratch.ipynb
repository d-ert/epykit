```python
import epykit
from epykit.io import read_samples
from epykit.core import MethylData
from epykit.stats.tests import calculate_diff_meth
```

```python
adata = epykit.io.read_samples(
    "samplesheet.csv",
    min_coverage=10,
    engine="duckdb"
)

# Optional: save to persistent Zarr storage
adata = epykit.io.read_samples(
    "samplesheet.csv",
    min_coverage=10,
    engine="duckdb",
    output="zarr",
    out_path="cohort.zarr"
)
```

```python
mdata = MethylData(adata)
mdata = mdata.filter_coverage(min_cov=10, max_cov=500)

if "context" in mdata.sites.columns:
    mdata = mdata.subset_context("CpG")

mdata = mdata.unite(how="intersect")
```

```python
results = calculate_diff_meth(
    mdata,
    treatment_col="group",
    test="auto",
    overdispersion=True,
    fdr_method="BH",
)

dmcs = results[results["qvalue"] < 0.05]
```
